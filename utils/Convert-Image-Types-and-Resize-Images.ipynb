```python
# Convert images from TIFF to PNG and Resize them to feed to the network

import os
from PIL import Image
import pdb

yourpath = r"C:\Users\kardoost\Desktop\Work-Folder\images\Inaki-3\video-frames\video-frames"

counter = 0
for root, dirs, files in os.walk(yourpath, topdown=False):
    for name in files:
        counter += 1
        print(os.path.join(root, name))
        if os.path.splitext(os.path.join(root, name))[1].lower() == ".tif":
            if os.path.isfile(os.path.splitext(os.path.join(root, name))[0] + ".png"):
                print("A png file already exists for {}".format(name))
            else:
                outfile = os.path.join(root, name)
                outfile = os.path.splitext(outfile)[0] + ".png"
                print(outfile)
                im = Image.open(os.path.join(root, name))
                print("Generating png for {}".format(name))
                size = 534, 534
                im.thumbnail(size, Image.ANTIALIAS)
                im.save(outfile, "PNG", quality=100)                
```
